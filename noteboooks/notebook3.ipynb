```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
main_csv = pd.read_csv("D:/(JSN)DataScience/KaggleProjects/StudentPerfromance/Students_Grading_Dataset.csv")
df = main_csv.copy()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=df, x='Age', hue='Stress_Level (1-10)')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Number of People at Each Stress Level per Age Group')
plt.legend(title='Stress Level')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=df, x='Department', y='Sleep_Hours_per_Night', hue='Age')
plt.xlabel('Department')
plt.ylabel('Sleep hours')
plt.title('Sleep hours by Department comparison')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(
    data=df,
    x='Stress_Level (1-10)',
    y='Total_Score',
    hue='Family_Income_Level'
)
plt.xlabel('Stress_Level (1-10)')
plt.ylabel('Score')
plt.title('Score')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data=df, x='Sleep_Hours_per_Night', y='Total_Score')
plt.xlabel('Sleep_Hours_per_Night')
plt.ylabel('Score')
plt.title('Score')
plt.tight_layout()
plt.show()
```

```python
# Create pairplot for key performance indicators
sns.pairplot(df[['Study_Hours_per_Week', 'Sleep_Hours_per_Night',
                'Stress_Level (1-10)', 'Total_Score', 'Grade']],
             hue='Grade', height=2.5)
plt.suptitle("Relationships Between Key Performance Indicators", y=1.02)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='Department', y='Total_Score')
plt.title("Distribution of Total Scores Across Departments")
plt.xlabel("Department")
plt.ylabel("Total Score")
plt.tight_layout()
plt.show()
```

```python
# Create correlation matrix for numerical variables
plt.figure(figsize=(14, 10))
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation = df[numeric_cols].corr()
mask = np.triu(correlation)
sns.heatmap(correlation, annot=True, fmt=".2f", cmap="coolwarm", mask=mask)
plt.title("Correlation Between Numerical Variables")
plt.tight_layout()
plt.show()
```
